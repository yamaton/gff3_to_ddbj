```python
import pandas as pd
# https://docs.google.com/spreadsheets/d/1qosakEKo-y9JjwUO_OFcmGCUfssxhbFAm5NXUAnT3eM/edit#gid=0
path = "./ddbj-feature-qualifier-matrix.xlsx"
```

```python
df_orig = pd.read_excel(path)
df_orig
```

```python
# select relevant part
df = df_orig.iloc[6:, 3:]
df
```

```python
# reset index and column names, and select only the content
df.index = df.iloc[:,0]
df = df.rename(columns=df.iloc[0])
df
```

```python
# Remove the header column and the header row
table = df.iloc[1:, 1:]
table
```

```python
# Get number of rows with proper (= not NaN) indices 
row_idx = next(i for i, x in enumerate(pd.isna(table.index)) if x)
row_idx

# Get number of columns with proper (= not NaN) heading 
col_idx = next(i for i, x in enumerate(pd.isna(table.columns)) if x)
col_idx

# Extract relevant table part
table_final = table.iloc[:row_idx, :col_idx]
table_final
```

```python
# Get mandatory feature-qualifier items as dictionary
mandatory = dict()
for name, row in table.iterrows():
    row_selected = row[row == "◎"]
    qualifiers = list(row_selected.index)
    if qualifiers:
        mandatory[name] = qualifiers

mandatory
```

```python
# Get optional feature-qualifier items as dictionary

optional = dict()
for name, row in table.iterrows():
    criteria = (row != "◎") & (~pd.isna(row))
    row_selected = row[criteria]
    qualifiers = list(row_selected.index)
    if qualifiers:
        optional[name] = qualifiers
optional
```

```python
# export the dictionaries as YAML / JSON

import collections
d = collections.defaultdict(dict)
for k, xs in optional.items():
    d[k]["optional_qualifiers"] = xs
for k, xs in mandatory.items():
    d[k]["mandatory_qualifiers"] = xs
d = dict(d)
```
